```python
import csv

CSV_List = None
# opening the CSV file
with open(f'{fname}.csv', mode ='r')as file:
   
  # reading the CSV file
  csvFile = csv.reader(file)
 
  # displaying the contents of the CSV file
  CSV_List = list(csvFile)
```

```python
#VISUALIZING THE SHAPE OF THE CSV
print(CSV_List[0])
features = int(CSV_List[0][0])
map_height = int(CSV_List[0][1])
map_width = int(CSV_List[0][2])
print(features, map_height, map_width)
```

```python
#PROCESSING THE CSV
ls = []
for i in CSV_List[1:]:
  t1 = []
  t2 = []
  for j in range(0,features * map_width,features):
    t2.append(list(map(float, i[j:j+features])))
  ls.append(t2)
```
